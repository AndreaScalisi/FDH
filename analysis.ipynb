# Analysis of the results

## Distribution of famous people among Paris arrondissements

```python
arr = [11, 1, 2, 4, 17, 18, 10, 8, 9, 13, 5, 7, 20, 19, 3, 16, 15, 14, 6, 12]
counts_arr = pd.DataFrame(index = arr)
counts_arr['1884'] = [3, 210, 45, 6, 66, 1, 10, 1010, 319, 8, 23, 862, 1, 2, 1, 192, 4, 1, 206, 1]
counts_arr['1884'] = counts_arr['1884'].map(lambda x: 100*x/counts_arr['1884'].sum())
counts_arr['1908'] = [2, 131, 17, 17, 478, 4, 2, 2394, 203, 2, 30, 1274, 0, 0, 0, 1250, 18, 7, 169, 10]
counts_arr['1908'] = counts_arr['1908'].map(lambda x: 100*x/counts_arr['1908'].sum())
counts_arr.sort_index(inplace = True)
```

```python
counts_arr.plot(kind = 'bar', width = 0.8, fontsize = 14)
plt.xlabel('Arrondissement', fontsize = 14)
plt.xticks(rotation = 'horizontal')
plt.ylabel('Proportion of famous people [%]', fontsize = 14)
plt.title('Distribution of famous people among Paris arrondissements in 1884 and 1908', fontsize = 20)
plt.legend(fontsize = 16)
plt.savefig('props_arr.png')
plt.show()
```
